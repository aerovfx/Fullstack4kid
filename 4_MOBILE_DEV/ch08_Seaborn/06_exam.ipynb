**Biểu đồ phân phối đa biến**: Thư viện này hỗ trợ các biểu đồ phân phối đa biến như `pairplot` và `jointplot`.

```python
import seaborn as sns # ✅ Import seaborn
import matplotlib.pyplot as plt # ✅ Import matplotlib
import pandas as pd  # ✅ Import pandas
from sklearn.datasets import load_iris # ✅ Import bộ dữ liệu iris

# Tải bộ dữ liệu iris
iris = load_iris() # ✅ Tải bộ dữ liệu iris
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names) # ✅ Tạo DataFrame từ iris.data
iris_df['species'] = iris.target # ✅ Thêm cột species vào DataFrame

# Biểu đồ pairplot
sns.pairplot(iris_df, hue='species') # ✅ Vẽ biểu đồ pairplot
plt.show() # ✅ Hiển thị biểu đồ

# Biểu đồ jointplot
sns.jointplot(x='sepal length (cm)', y='sepal width (cm)', data=iris_df, kind='scatter') # ✅ Vẽ biểu đồ jointplot
plt.show()  # ✅ Hiển thị biểu đồ
```
