```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# To mt tp d liu n gin bng NumPy
x = np.linspace(0, 10, 100)
y = np.sin(x)

# To mt DataFrame bng Pandas
df = pd.DataFrame({'x': x, 'y': y})

# V d liu bng Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(df['x'], df['y'], label='sin(x)')

# Thm cc ng nt t cho trc x
for xc in np.arange(0, 11, 1):
    plt.axvline(x=xc, color='r', linestyle='--', linewidth=0.5)

plt.title('Biu  n gin s dng Matplotlib, NumPy v Pandas')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```
