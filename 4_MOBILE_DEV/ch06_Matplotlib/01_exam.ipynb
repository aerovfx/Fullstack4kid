```python
import matplotlib.pyplot as plt
"""
V  Th a Dng vi Matplotlib

Chng trnh ny minh ha cch v cc loi  th khc nhau s dng th vin Matplotlib bao gm:
-  th ng (Line plot)
-  th ct (Bar plot)
-  th phn tn (Scatter plot)
- Biu  histogram (Histogram)
-  th 3D (3D plot)

Cc phn chnh ca chng trnh:
1. Khi to d liu cho cc  th.
2. V  th ng vi hm sin(x) v cos(x).
3. V  th ct vi cc danh mc v gi tr tng ng.
4. V  th phn tn vi cc im ngu nhin.
5. V biu  histogram vi d liu ngu nhin.
6. V  th 3D vi hm sin ca khong cch t gc ta .

Cc th vin s dng:
- matplotlib.pyplot: Th vin chnh  v  th.
- numpy: Th vin  x l mng v to d liu ngu nhin.
- mpl_toolkits.mplot3d: Th vin  v  th 3D.
"""
import numpy as np

# Data for plotting
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
data = np.random.randn(1000)

# Line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]

plt.figure(figsize=(10, 6))
plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Scatter plot
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)

plt.figure(figsize=(10, 6))
plt.scatter(x_scatter, y_scatter)
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, alpha=0.7, color='g')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# 3D plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
x_3d = np.linspace(-5, 5, 100)
y_3d = np.linspace(-5, 5, 100)
x_3d, y_3d = np.meshgrid(x_3d, y_3d)
z_3d = np.sin(np.sqrt(x_3d**2 + y_3d**2))

ax.plot_surface(x_3d, y_3d, z_3d, cmap='viridis')
ax.set_title('3D Plot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```
